```python
import catboost
from catboost import CatBoostClassifier
```

```python
params = {'iterations':5000,
        'learning_rate':0.01,
        'depth':8,
        'eval_metric':'Accuracy',
        'verbose':200,
        'od_type':"Iter", # overfit detector
        'od_wait':1000, # most recent best iteration to wait before stopping
        'random_seed': 2,
          'loss_function':'MultiClass'
          }

cat_model = CatBoostClassifier(**params)
cat_model.fit(x_train, y_train,   
          eval_set=(x_test, y_test), 
          use_best_model=False, # True if we don't want to save trees created after iteration with the best validation score
          plot=True);
```

```python
cat_model.best_score_
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_test,cat_model.predict(x_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, cat_model.predict(x_test))
```

# Artificial Neural Network

```python
Dummies = pd.get_dummies(Data.iloc[:,-1])
Data = Data.drop(['label'],axis=1)
Data = pd.concat((Data,Dummies),axis=1)
```

```python
Y = Data.iloc[:,24:]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y.values, test_size=0.1)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import models, layers, optimizers
from matplotlib import pyplot
plt.style.use('ggplot')
from livelossplot import PlotLossesKeras
```

```python
import time
# starting time
start = time.time()
model = models.Sequential()
model.add(layers.Dense(100, input_dim = len(X.columns), activation='relu'))
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(len(Y.columns), activation='softmax'))
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
history = model.fit(X_train, Y_train, epochs=500, batch_size=300,validation_data=(X_test, Y_test))
end = time.time()
print(f"Runtime of the program is {end - start}")
```

```python
model.save("ANN_Model")
```

```python
import matplotlib.pylab as plt
from matplotlib.pyplot import figure
figure(figsize=(8, 6))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.savefig('Loss',dpi=100,bbox_inches='tight')
plt.show()
```

```python
figure(figsize=(8, 6))
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.savefig('Accuracy',dpi=100,bbox_inches='tight')
plt.show()
```

```python
model.evaluate(X_test,Y_test,verbose=0)
```

```python
y_testing = np.argmax(model.predict(X_test),axis=1)
y_actual = np.argmax(Y_test.values,axis=1)
```
